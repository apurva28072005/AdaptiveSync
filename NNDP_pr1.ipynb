```python
import numpy as np
v1=np.array([10,20,30])
v2=np.array([15,25,20])
print("Vector 1:",v1)
print("Vector 2:",v2)
result1=v1+v2
print("Result :",result1)
print()
print("2*2 Matrix")
m1=([10,20],[40,50])
m2=([15,25],[55,65])
print("Matrix 1:",m1)
print("Matrix 2:",m2)
result2=np.matmul(m1,m2)
print("Result :",result2)
print()
print("3*3 Matrix")
m3=([10,20,30],[40,50,60],[70,80,90])
m4=([15,25,35],[55,65,75],[85,95,105])
print("Matrix 3:",m3)
print("Matrix 4:",m4)
result3=np.matmul(m3,m4)
print("Result :",result3)
print()
print("3*2-2*3 Matrix")
m5=([10,20,30],[40,50,60])
m6=([15,25],[65,75],[95,105])
print("Matrix 5:",m5)
print("Matrix 6:",m6)
result4=np.matmul(m5,m6)
print("Result :",result4)



```
